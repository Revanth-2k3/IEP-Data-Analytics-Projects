```python
# Drop irrelevant columns
data_cleaned = data.drop(columns=["EmployeeCount", "StandardHours", "Over18", "EmployeeNumber"])
```

```python
# Summary statistics of numerical columns
data_cleaned.describe()
```

```python
# Distribution of the target variable
attrition_distribution = data['Attrition'].value_counts(normalize=True) * 100
print("Attrition Distribution:\n", attrition_distribution)
```

```python
# Distribution of categorical variables
categorical_columns = data.select_dtypes(include=['object']).columns
```

```python
for col in categorical_columns:
    print(f"Distribution of {col}:\n", data[col].value_counts())
    sns.countplot(x=col, hue='Attrition', data=data)
    plt.title(f'Attrition by {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Distribution of numerical variables
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
```

```python
for col in numerical_columns:
    sns.boxplot(x='Attrition', y=col, data=data)
    plt.title(f'Attrition by {col}')
    plt.show()
```

```python
# Pairplot for key numerical variables
sns.pairplot(data[['Age', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany', 'Attrition']], hue='Attrition')
plt.show()
```

```python
# Create a new feature: Average Income per Year
data['AvgIncomePerYear'] = data['MonthlyIncome'] * 12

# Create a new feature: Years since last promotion
data['YearsSinceLastPromotion'] = data['YearsAtCompany'] - data['YearsInCurrentRole']

# Display the new features
data[['AvgIncomePerYear', 'YearsSinceLastPromotion']].head()
```

```python
sns.countplot(x='Attrition', data=data)
plt.title('Attrition Distribution')
plt.show()
```

```python
# Identify categorical columns
categorical_cols = data_cleaned.select_dtypes(include=["object"]).columns
```

```python
# Convert categorical variables using one-hot encoding
data_encoded = pd.get_dummies(data_cleaned, columns=categorical_cols, drop_first=True)
```

```python
# Compute correlation matrix again
corr_matrix = data_encoded.corr()
```

```python
# Plot heatmap
plt.figure(figsize=(14,7))
sns.heatmap(corr_matrix, cmap="coolwarm", annot=False, fmt=".2f")
plt.title("Correlation Heatmap (Encoded Data)")
plt.show()
```
