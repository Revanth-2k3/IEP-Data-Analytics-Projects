```python
#  Creating a Binary Classification Target Variable
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data = pd.DataFrame(data)
```

```python
# Create a binary classification target variable: 'price_category' based on price
threshold = data['price'].median()  # You can change the threshold here
data['price_category'] = np.where(data['price'] > threshold, 1, 0)
```

```python
# Features (X) and target (y)
X = data.drop(['price', 'price_category'], axis=1)
y = data['price_category']
```

```python
# Splitting the Data into Training and Testing Sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Standardizing the Features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Initializing and Training the Logistic Regression Model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
# Making Predictions
y_pred = model.predict(X_test)
```

```python
# Evaluating the Model's Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.2f}%')
```
